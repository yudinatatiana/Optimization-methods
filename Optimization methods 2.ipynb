```python
def objective_function(U1, U1Z):
    #Целевая функция
    #U1 - варьируемый параметр, U1Z - заданный параметр
    return U1**2 - 2 * U1 * U1Z + U1Z**2
```

```python
def objective_function_derivative(U1, U1Z):
    #Производная целевой функции
    return 2 * U1 - 2 * U1Z
```

```python
def main(U1, U1Z_1, U2Z_1, U3Z_1):
    print('Введите номер метода оптимизации для запуска')
    print('Методы первого порядка:')
    print('1 - Метод градиентного спуска')
    print('2 - Метод градиентного спуска с оптимальным шагом')
    print('3 - Метод сопряжённых градиентов')
    print('Методы второго порядка:')
    print('4 - Метод Ньютона')
    number = int(input())
    if number == 1:
        gradient_descent(U1, U1Z_1, U2Z_1, U3Z_1)
    if number == 4:
        newton_method(U1, U1Z_1, U2Z_1, U3Z_1)
    if number == 3:
        conjugate_gradient_method(U1, U1Z_1, U2Z_1, U3Z_1)
    if number == 2:
        gradient_descent_optimal_step(U1, U1Z_1, U2Z_1, U3Z_1)
```

```python
def gradient_descent(U1, U1Z, U2Z_1, U3Z_1):
    #Метод градиентного спуска
    print('Введите максимальное количесто итераций')
    max_iter = int(input())
    i = 0
    eps = 0.001
    kof = 0.1
    prev_loss = objective_function(U1, U1Z)
    parametr = U1
    while i < max_iter:
        parametr -= kof * objective_function_derivative(parametr, U1Z)
        loss = objective_function(parametr, U1Z)
        print('Шаг', i)
        print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
        print('Целевая функция', loss)
        if abs(loss-prev_loss) < eps:
            break
        prev_loss = loss
        i += 1
```

```python
def gradient_descent_optimal_step(U1, U1Z, U2Z_1, U3Z_1):
    #Метод градиентного спуска с оптимальным шагом
    print('Введите максимальное количесто итераций')
    max_iter = int(input())
    i = 0
    eps = 0.001
    prev_loss = objective_function(U1, U1Z)
    parametr = U1
    while i < max_iter:
        kof = (U1Z - parametr) / objective_function_derivative(parametr, U1Z)
        parametr += kof * objective_function_derivative(parametr, U1Z)
        loss = objective_function(parametr, U1Z)
        print('Шаг', i)
        print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
        print('Целевая функция', loss)
        if abs(loss-prev_loss) < eps:
            break
        prev_loss = loss
        i += 1
```

```python
def newton_method(U1, U1Z, U2Z_1, U3Z_1):
    #Метод Ньютона
    print('Введите максимальное количесто итераций')
    max_iter = int(input())
    i = 0
    eps = 0.001
    prev_loss = objective_function(U1, U1Z)
    parametr = U1
    while i < max_iter:
        parametr -= objective_function(parametr, U1Z) / objective_function_derivative(parametr, U1Z)
        loss = objective_function(parametr, U1Z)
        print('Шаг', i)
        print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
        print('Целевая функция', loss)
        if abs(loss - prev_loss) < eps:
            break
        prev_loss = loss
        i += 1
```

```python
def conjugate_gradient_method(U1, U1Z, U2Z_1, U3Z_1):
    #Метод сопряжённых градиентов
    print('Введите максимальное количесто итераций')
    max_iter = int(input())
    i = 0
    eps = 0.001
    parametr = U1
    while i < max_iter:
        if i == 0:
            loss = objective_function(parametr, U1Z)
            grad = objective_function_derivative(parametr, U1Z)
            norm = (grad**2)**0.5
            i += 1
            print('Шаг', i)
            print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
            print('Целевая функция', loss)
        elif i == 1:
            t = (U1Z - parametr) / grad
            parametr -= t - objective_function_derivative(parametr, U1Z)
            loss = objective_function(parametr, U1Z)
            grad = objective_function_derivative(parametr, U1Z)
            norm = (grad**2)**0.5
            d = -grad
            i += 1
            print('Шаг', i)
            print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
            print('Целевая функция', loss)
        else:
            new_grad = objective_function_derivative(parametr, U1Z)
            new_norm = (new_grad**2)**0.5
            b = new_norm**2 / norm**2
            dk = -new_grad + b * d
            #вычислить t
            t = (U1Z - parametr) / grad
            parametr -= t * dk
            norm = new_norm
            d = dk
            loss = objective_function(parametr, U1Z)
            print('Шаг', i)
            print('Новый параметр:', parametr, 'Расчитанный параметр', U1Z)
            print('Целевая функция', loss)
            i += 1
            if (norm < eps) or (loss < eps):
                break
```
